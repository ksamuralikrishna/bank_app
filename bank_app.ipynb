```python
import logging
import sqlite3
import random

logging.basicConfig(level=logging.DEBUG,
                    filename="applog.txt",
format="%(asctime)s=>%(levelname)s==>%(message)s")

logging.debug("data base is created")
logging.debug("bank_app table is created")

con = sqlite3.connect("mydb.db")
cur=con.cursor()
cur.execute("create table bank_app(id int,name varchar,age int,addr varchar,type varchar,loan varchar,date int(8),status varchar)")
con.commit()

q="insert into bank_app values(001123,'vishnu',43,'Guntur','gold','personal',06081999,'running')"
cur.execute(q)
con.commit()

logging.info("User create dtabse sucessfully ")
logging.info("User create bank_app sucessfully ")

logging.debug("user taken random values")

names=["kurama","matya","varaha","jayaram","krishna","budha","murali"]
age=['45','43','47','48']
addr=['Vizag','chittur','Vijayawada','Guntur','Hyderabad']
types=['gold','silver','normal']
loans=['personal','houseloan','jumboloan']
dates=['090917','120917','310917','040817']
status=['running','close']
for i in range(1124,1172):
    q="insert into bank_app values(%s,'%s','%s','%s',%s,'%s','%s','%s')" %(i,random.choice(names),random.choice(age),random.choice(addr),random.choice(types),random.choice(loans),random.choice(dates),random.choice(status)) 
    cur.execute(q)
con.commit()

logging.info("User taken values sucessfully ")


logging.debug("user print query")
q="select * from bank_app"
cur.execute(q)
con.commit()

logging.info("print query sucessfully ")


logging.debug("user update query")
q="update bank_app set name='tara' where id='1123'"
cur.execute(q)
con.commit()

logging.info("User update sucessfully ")


logging.debug("check delete query")
q="delete bank_app where id='1172'"
cur.execute(q)
con.commit()

logging.info("delete row sucessfull ")
```
